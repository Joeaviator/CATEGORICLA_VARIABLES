```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Titanic Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['PassengerId'],axis=1,inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
#mode gender
df['Gender'].value_counts()
```

```python
sns.countplot(x=(df['Gender']), palette='turbo')
plt.title("Gender division")
```

```python
#Median gender
gender_categories=["male","female"]
df['Gender']=pd.Categorical(df['Gender'],gender_categories,ordered=True)
median_index=np.median(df['Gender'].cat.codes)
median_gender=gender_categories[int(median_index)]
print(f"Median Gender: {median_gender}")
```

```python
df['Embarked'].value_counts()
```

```python
sns.countplot(x=(df['Embarked']), palette='turbo')
plt.title("Embarked division")
```

```python
Embarked_Categories = ["C" , "S"]
df['Embarked']=pd.Categorical(df['Embarked'],Embarked_Categories,ordered=True)
median_index=np.median(df['Embarked'].cat.codes)
median_Embarked=Embarked_Categories[int(median_index)]
print(f"Median Embarked: {median_Embarked}")
```

```python
#Piechart Of Survived
sns.countplot(x=(df['Survived']) , palette='rainbow')
plt.title("Survived")
```

```python
#Piechart Of Survived
survived_counts=df['Survived'].value_counts()
label=["Did not survive", "Survive"]
plt.pie(survived_counts,labels=label,autopct='%1.1f%%')
plt.title("Survived")
plt.show()
```

```python
Embarked_counts=df['Embarked'].value_counts()
label=["Cherbourg", "Southampton"]
colors=['turquoise','red']
plt.pie(Embarked_counts,labels=label,autopct='%1.1f%%',colors=colors)
plt.title("Embarked")
plt.show()
```
